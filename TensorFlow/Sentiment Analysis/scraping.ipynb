```python
from google_play_scraper import app, reviews, Sort, reviews_all
import csv
```

```python
ca_scraps = reviews_all(
    'com.openai.chatgpt',
    lang='en',
    country='ca',
    sort=Sort.MOST_RELEVANT,
    count=3000
)

us_scraps = reviews_all(
    'com.openai.chatgpt',
    lang='en',
    country='us',
    sort=Sort.MOST_RELEVANT,
    count=3000
)

au_scraps = reviews_all(
    'com.openai.chatgpt',
    lang='en',
    country='au',
    sort=Sort.MOST_RELEVANT,
    count=3000
)

in_scraps = reviews_all(
    'com.openai.chatgpt',
    lang='en',
    country='in',
    sort=Sort.MOST_RELEVANT,
    count=3000
)

gb_scraps = reviews_all(
    'com.openai.chatgpt',
    lang='en',
    country='gb',
    sort=Sort.MOST_RELEVANT,
    count=3000
)
```

```python
scraps_dict = {
    'ca': ca_scraps,
    'us': us_scraps,
    'au': au_scraps,
    'in': in_scraps,
    'gb': gb_scraps
}

for country, data in scraps_dict.items():
    filename = f'chatgpt_review_{country}.csv'
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['content'])
        for review in data:
            writer.writerow([review['content']])
    print(f'Successfully saved {len(data)} reviews to {filename}')
```
